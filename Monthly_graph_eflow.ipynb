```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
H_flow=['May ','Jun','Jul','Aug', 'Sep','Oct']
L_flow=['Nov','Dec','Jan','Feb','Mar','Apr']
river_name=['Balu River SW7.5', 'Buriganga River SW42', 'Dhaleshawri River SW68A', 'Shitalakhya River SW179', 'Turag River SW301','Turag River SW302', 'Tongi Canal SW299']
```

```python
river_name[6]
```

```python
plt.rcParams["figure.figsize"] = [6, 4]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots(1,1, squeeze= True)

for n in np.arange(0,7): 
  df= pd.read_excel('e-flow_Part1.xlsx',sheet_name=n)
  df.drop(['EFR'], axis=1, inplace=True)
  df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
  filt=df['Month'].isin(H_flow)
  H_flow_df=df[filt]
  filt1=df['Month'].isin(L_flow)
  L_flow_df=df[filt1]
  H_flow_df = H_flow_df.set_index(H_flow_df['Month'])
  L_flow_df = L_flow_df.set_index(L_flow_df['Month']).loc[L_flow]
  ax.plot("Mean Monthly Flow (cumec)",data=H_flow_df,label='Monthly Mean Flow (cumec)',color='salmon')
  ax.plot("Tennant Method",data=H_flow_df,label='Tennant Method',color='green')
  ax.plot("Flow Duration Method",data=H_flow_df,label='Flow Duration Method',color='steelblue')
  ax.set_title(river_name[n],fontweight = 'semibold',fontsize = 20)
  ax.set_xlabel('Month',fontweight = 'semibold',fontsize = 16)
  ax.set_ylabel('Flow (cumec)',fontweight = 'semibold', fontsize = 16)
  ax.grid(visible = True, axis = 'y',  alpha = 0.6)
  ax.xaxis.set_tick_params(labelsize=14)
  ax.yaxis.set_tick_params(labelsize=14)
  # fig.legend(handles, labels, loc=(1,1))
  plt.tight_layout()
  plt.savefig(river_name[n]+"_High_E-Flow_Monthly_GRAPH.tiff", dpi=500)
  plt.show()
```
