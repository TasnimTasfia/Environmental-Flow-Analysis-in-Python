```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
river_name=['Balu River SW7.5', 'Buriganga River SW42', 'Dhaleshawri River SW68A', 'Shitalakhya River SW179', 'Turag River SW301','Turag River SW302', 'Tongi Canal SW299']
```

```python
plt.rcParams["figure.figsize"] = [6, 4]
plt.rcParams["figure.autolayout"] = True

for n in np.arange(0,7): 
  df1= pd.read_excel('prediction_graph2.xlsx',sheet_name=n)
  fig, ax = plt.subplots(1,1, squeeze= True)

  ax.plot(df1['Date'],df1['predicted'],label='Predicted Flow')
  ax.plot(df1['Date'],df1['MMF'],label='Mean Monthly Flow')
  # plt.yticks(np.arange(50,300,25))
  plt.title(river_name[n],fontweight = 'semibold',fontsize = 20)
  plt.xlabel('Time(year)',fontweight = 'semibold',fontsize = 16)
  plt.ylabel('Flow (cumec)',fontweight = 'semibold',fontsize = 16)
  plt.grid(visible = True, axis = 'y',  alpha = 0.6)
  #plt.legend()
  ax.xaxis.set
  ax.xaxis.set_tick_params(labelsize=14)
  ax.yaxis.set_tick_params(labelsize=14)
  fig.autofmt_xdate()
  plt.tight_layout()
  plt.savefig(river_name[n]+"predicted_.tiff", dpi=500)
  plt.show()
```
